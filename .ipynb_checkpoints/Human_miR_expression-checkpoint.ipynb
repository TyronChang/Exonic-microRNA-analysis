### Webscraping tables from MirGeneDB 2.1

```python
r=requests.get('https://mirgenedb.org/browse/hsa')
```

```python
webpage=bs(r.content,'html.parser')
```

```python
# webpage.find('tbody').find_all('tr')
```

### Generate row names 

```python
table_rows=webpage.find('tbody').find_all('tr')
row_name=[]
for tr in table_rows:
    td=tr.find_all('td')
    rows=[rows.text for rows in td]
    row_name.append(rows)
row_names=[[row[0],row[1].strip()] for row in row_name]
# print(row_names)
```

### Generate column names 

```python
columns=webpage.find('thead').find_all('th')
col_name=[column.text for column in columns]
col_names=[col_name[5].replace("\n    ▲",""),col_name[6].replace("\n   ","")]
col_names=[col_names[0].strip(),col_names[1].strip()]
col_names=[col_names[0].replace(" ","_"),col_names[1].replace(" ","_")]
print(col_names)
```

```python
hsa_miRgene_db=pd.DataFrame(row_names,columns=col_names)
hsa_miRgene_db
```

```python
human_miR_df['MirGeneDB_ID']=human_miR_df['MirGeneDB_ID'].str.replace("_pri","")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame with zero and negative values
data = {
    'Column1': [1, 2, 0, -8],
    'Column2': [10, 20, -30, 40]
}
human_miR_df_merge = pd.DataFrame(data)

# Replace zeros with a small positive number and handle negative values
small_value = 1e-9
human_miR_df_merge = human_miR_df_merge.replace(0, small_value)
human_miR_df_merge[human_miR_df_merge < 0] = np.nan

# Apply log2 transformation
log_df = np.log2(human_miR_df_merge)

# Replace inf and -inf with NaN
log_df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows or columns with NaN values
log_df.dropna(axis=0, inplace=True)
log_df.dropna(axis=1, inplace=True)

# Create clustermap
fig = sns.clustermap(log_df, xticklabels=True, yticklabels=True, linewidths=1,
                     cbar=True, cbar_kws={'label': u'Log\u2082TPM'}, cmap='coolwarm',
                     vmin=0, vmax=10, figsize=(10, 10))

plt.show()
```
